```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rhq = pd.read_sas("../data/raw/P_RHQ.XPT", format="xport")

print("="*80)
print("NHANES REPRODUCTIVE HEALTH MODULE - DETAILED EXPLORATION")
print("="*80)
print(f"\nDataset: {rhq.shape[0]:,} participants × {rhq.shape[1]} variables")

# Key variables we found in the search
key_vars = {
    'SEQN': 'Participant ID',
    'RHQ131': 'Ever been pregnant?',
    'RHQ160': 'Number of pregnancies',
    'RHQ162': 'Gestational diabetes?',
    'RHQ172': 'Baby ≥9 lbs (macrosomia)?',
    'RHD167': 'Total deliveries',
    'RHQ171': 'Live births',
    'RHD180': 'Age at first live birth',
    'RHD190': 'Age at last live birth',
    'RHD018': 'Age at menarche (months)',
    'RHQ010': 'Age at first period (years)',
    'RHQ060': 'Age at last period',
    'RHQ031': 'Regular periods past 12mo?',
    'RHD280': 'Had hysterectomy?',
    'RHQ305': 'Both ovaries removed?',
    'RHQ540': 'Ever used hormones?',
    'RHQ074': 'Tried ≥1yr to conceive?',
    'RHQ076': 'Saw DR for infertility?',
}

print("\n" + "="*80)
print("KEY REPRODUCTIVE VARIABLES - AVAILABILITY CHECK")
print("="*80)

found = []
missing = []

for var, desc in key_vars.items():
    if var in rhq.columns:
        n_valid = rhq[var].notna().sum()
        pct = (n_valid / len(rhq)) * 100
        found.append((var, desc, n_valid, pct))
        print(f"Foound: {var:10s} | {desc:40s} | {n_valid:5,} ({pct:5.1f}%)")
    else:
        missing.append((var, desc))
        print(f"Missing: {var:10s} | {desc:40s} | NOT FOUND")

print(f"\n Found: {len(found)}/{len(key_vars)} variables")

# Now look at actual values for the CRITICAL variables
print("\n" + "="*80)
print("CRITICAL VARIABLES - VALUE DISTRIBUTIONS")
print("="*80)

critical_vars = ['RHQ131', 'RHQ162', 'RHQ172', 'RHD167', 'RHQ171', 'RHQ060', 'RHQ031']

for var in critical_vars:
    if var in rhq.columns:
        print(f"\n{var}: {key_vars[var]}")
        print("-" * 80)
        vc = rhq[var].value_counts(dropna=False).sort_index()
        print(vc)
        
        # Show some stats if numeric
        if rhq[var].dtype in ['float64', 'int64']:
            valid_data = rhq[var].dropna()
            if len(valid_data) > 0:
                print(f"\nStats: Min={valid_data.min():.1f}, "
                      f"Mean={valid_data.mean():.1f}, "
                      f"Max={valid_data.max():.1f}")
```
